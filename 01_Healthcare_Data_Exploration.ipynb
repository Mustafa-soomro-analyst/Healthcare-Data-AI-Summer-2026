```python
import pandas as pd
medical_data = {
    'Patient_ID': [101,102,103,104,105,106,107,108],
    'Age': [45,22,71,34,65,58,19,82],
    'Gender': ['M','F','M','F','M','F','M','F'],
    'Condition': ['Diabetes', 'Influenza', 'Heart Disease', 'Diabetes','Heart Disease', 'Ínfluenza', 'Apendicitis', 'Heart Disease'],
    'Admission_Type':['Elective', 'Emergency', 'Emergency', 'Urgent', 'Emergency', 'Urgent', 'Emergency', 'Elective'],
    'Billing_Amount_USD': [1200, 450, 8500, 1500, 9200, 600, 3100, 7800]
}

df = pd.DataFrame(medical_data)

print("--- Healthcare Dataset Preview---")
print(df.head())

print("\n--- Basic Medical Data Insights---")
print(f"Average Patient Age: {df['Age'].mean()} years old")
print(f"Total Hospital Revenue: ${df['Billing_Amount_USD'].sum()}")

print("\n--- Patient Demographics---")
print(df['Gender'].value_counts())

print("\n--- Patient Conditions---")
print(df['Condition'].value_counts())

print("\n--- Admission Types---")
print(df['Admission_Type'].value_counts())

```

```python
average_billing_per_condition = df.groupby('Condition')['Billing_Amount_USD'].mean()
print("\n--- Average Billing Amount per Condition---")
print(average_billing_per_condition)

emergrncy_patients = df[df['Admission_Type'] == 'Emergency']
print("\n--- Emergency Patients---")
print(emergrncy_patients[['Patient_ID', 'Age', 'Condition']])

```

```python
import pandas as pd


dataset_url = "https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv"


df = pd.read_csv(dataset_url)


print("--- Dataset Successfully Loaded! ---")
print(f"Total Patient Records: {df.shape[0]}")
print(f"Total Metrics Collected: {df.shape[1]}")


print("\n--- Clinical Attributes List ---")
print(df.columns.tolist())


print("\n--- First 5 Patient Rows ---")
print(df.head())
```

```python
#filtering high cholestrol patients

high_risk_patients = df[(df['age'] > 50) & (df['chol'] > 240)]

num_high_risk = high_risk_patients.shape[0]
print(f"Number of Patients over 50 with High Cholesterol: {num_high_risk}")

max_chol = high_risk_patients['chol'].max()
print(f"Highest individulal cholesterol level in this record: {max_chol} mg/dL")
```
